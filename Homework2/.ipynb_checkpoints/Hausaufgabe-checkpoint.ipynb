```python
import numpy as np
import torch
import torchvision
import imageio
import matplotlib.pyplot as plt
```

```python
# 10x10 Tensor
d = torch.tensor(np.arange(0,100).reshape([10,10]))
print(d)
```

```python
#  Transponierte des Vektors
transpose_t = d.transpose(0,1)
print(transpose_t)
print(id(transpose_t.untyped_storage()) == id(d.untyped_storage()))
```

```python
#9x9 Untermatrix, 1.Zeile und Spalte streichen
c = d[1:,1:]
print(c.stride())
print(c.size())
print(c.storage_offset())
```

```python
# inverse Matrix
einheit = torch.tensor(np.identity(10))

inv = torch.inverse(einheit)
print(inv*einheit)
```

```python
# Bildausschnitt mit niedriger Auflsung
img_arr = imageio.v2.imread('Cat2.jpg')

img = torch.from_numpy(img_arr)
#plt.imshow(img)

print(img.stride())
t_h = torch.as_strided(img, (500,512,3), (6144, 6, 1))
t = torch.as_strided(img, (300,350,3), (6144, 6, 1))

plt.imshow(t)
```
